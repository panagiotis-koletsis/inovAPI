```python
import docker

client = docker.from_env()

container_name = "fastapi-upload-app"  # your container name
files_to_delete = ["test.txt", "test1.txt"]  # paths inside the container

container = client.containers.get(container_name)

for file_path in files_to_delete:
    exit_code, output = container.exec_run(f"rm -f {file_path}")
    if exit_code == 0:
        print(f"Deleted {file_path}")
    else:
        print(f"Failed to delete {file_path}: {output.decode()}")
```
